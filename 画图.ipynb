```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("./2024_MCM-ICM_Problems/Problem_D_Great_Lakes.xlsx", sheet_name=0)
```

```python
data.head()
```

```python
data.index = data.Year
```

```python
data.drop(labels='Year', axis=1, inplace=True)
```

```python
data = data.T
data.head()
```

```python
# data = data.T
data = data.drop(labels=[2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2022],
                 axis=1,)
```

```python
data
```

```python
data.head()
```

```python

fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 10))

for i, column in enumerate(data.columns):
    row = i // 4  # calculate the row index
    col = i % 4   # calculate the column index
    
    axes[row, col].plot(data.index, data[column], marker='o', linestyle='-', linewidth=2)
    axes[row, col].set_xlabel('Year', fontsize=12)
    axes[row, col].set_ylabel('Value', fontsize=12)
    axes[row, col].set_title(f'Line Graph for {column}', fontsize=14)
    axes[row, col].legend([column], loc='best', fontsize=10)

plt.tight_layout()
plt.show()

for i, column in enumerate(data.columns):
    axes[i].plot(data.index, data[column], marker='o', linestyle='-', linewidth=2)
    axes[i].set_xlabel('Year', fontsize=12)
    axes[i].set_ylabel('Value', fontsize=12)
    axes[i].set_title(f'Flow for {column}', fontsize=14)
    axes[i].legend([column], loc='best', fontsize=10)

plt.tight_layout()
plt.show()

for i, column in enumerate(data.columns):
    row = i // 4  # calculate the row index
    col = i % 4   # calculate the column index
    
    axes[row, col].plot(data.index, data[column])
    axes[row, col].set_xlabel('Year')
    axes[row, col].set_ylabel('Value')
    axes[row, col].set_title(f'Flow for {column}')
    axes[row, col].legend([column], loc='best')

plt.tight_layout()
plt.show()

for i, column in enumerate(data.columns):
    axes[i].plot(data.index, data[column])
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Value')
    axes[i].set_title(f'Flow for {column}')
    axes[i].legend([column], loc='best')

plt.tight_layout()
plt.show()
```
